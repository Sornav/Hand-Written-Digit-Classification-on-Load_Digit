```python
from sklearn import datasets 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
dataset=datasets.load_digits()##importing load_digits dataset
print(dataset.keys())
print(dataset.DESCR)
```

```python
##displaying 1st 10 digits in the dataset
for i in range(10):
    some_digit = dataset.data[i]
    some_digit=some_digit.reshape(8,8)
    plt.imshow(some_digit, cmap=matplotlib.cm.binary)
    plt.axis("off")
    plt.show()
```

Getting the values from the dataset,we create the target and feature array or in simple words dependent and independent variable.

```python
x=dataset.data
y=dataset.target
cols=len(x[0])
k=np.insert(x,cols,y,axis=1)##combining the individual features into a single array
k
```

So it is clear that for each row in x we have a coresponding digit in y

*basically we are finding the nearest neighbors(closest distance) of new_data in training_data

*1st we need to find distance of new_data from each tarining_data

*we need to store the distance and the sort the data in the training_data according to the distance from new_data

*for example if the distance of test_row from 8th data of the train is lowest so it comes 1st in NN..and so on



So let us now build our own KNN model from scratch

```python
## Find Euclidean Distance
def Euclidean_Distance(row1, row2):##accepts 2 single rows
    distance = 0
    for i in range(len(row1)-1):
        distance += (row1[i] - row2[i])**2
    return distance**0.5


def Get_NN(train, test_row, k = 3):##accept a single test_row only ##gets nearest neighbors for a single test row in the train
    distance = []
    data = []
    for i in train:
        dist = Euclidean_Distance(test_row, i)
        distance.append(dist)
        data.append(i)
    ## convert distance and data in nump array
    distance = np.array(distance)
    data = np.array(data)
    "*** we are finding index number of min distance  ***"
    index_dist = distance.argsort()##this sorts the distance array in ascending order
    "** arange Data acco. to index_dist **"
    data = data[index_dist]
    "** find the Nearest Neighbors with slicing on data acco. to K value  **"
    neighbors = data[:k]
    return neighbors


def Predict_Classes(train, test_row, k):##used to predict the exact class of the data(either 0 or 1 or 2)##accept single test_row
    Neighbors = Get_NN(train, test_row, k)
    classes = []
    for i in Neighbors:
        classes.append(i[-1])#######################appending the classes(target) for each data
    prediction = max(classes, key = classes.count)##returning value with max count
    return prediction


def Predict(Train, test, k = 3):##since predict classes works with only a sinle test_row so we use a loop for each element in the test_row
    y_predict = []
    for i in test:
                   prediction = Predict_Classes(Train, i, k)
                   y_predict.append(prediction)
    return np.array(y_predict)


def Evaluate(y_true, y_pred):
    n_correct = 0
    for i,j in zip(y_true, y_pred):
        if i == j:
            n_correct += 1
    return n_correct/len(y_true)*100
```

let us now split our data,for simplicity we manually split it

```python
##by manually spliting the data
trainX=k[0:600]
testX=k[:70]
```

```python
y_pred = Predict(trainX,testX,6)
print(y_pred)##predicted result
plt.plot(y_pred)
plt.show()
```

```python
y_true = k[:70,-1]
print(y_true)##true
plt.plot(y_true)
plt.show()
```

```python
plt.plot(y_true)
plt.plot(y_pred)
plt.show()
```

```python
Evaluate(y_pred, y_true)##accuracy
```

## Let us Use sklearn


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
x = k[:,:-1]
y = k[:, -1] 
##Let us implement our own train_test_split
def Train_Test_Split(x,y=[0],test_size=0.5,random_state=None):
    n=len(x)
    if random_state:
        np.random.seed(random_state)##fixes random number
    if len(y)>2:##check y is empty or not
        if len(x)==len(y):
            shuffle_index=np.random.permutation(n)
            x=x[shuffle_index]
            y=y[shuffle_index]
            test_rows=round(len(x)*test_size)
            trainX,testX=x[test_rows:],x[:test_rows]
            trainY,testY=y[test_rows:],y[:test_rows]
            return trainX,testX,trainY,testY
        else:##if length and x and y is different generate error
            raise ValueError("Found input variable with inconsistent data")
            
    else:##spltitng only x because y not present
        x=x[shuffle_index]
        
        
        trainX,testX=x[test_rows:],x[:test_rows]
        return trainX,testX
trainX,testX,trainY,testY=Train_Test_Split(x,y,test_size=0.2,random_state=12)
    
    
KNN_Class = KNeighborsClassifier(n_neighbors=6)
KNN_Class = KNN_Class.fit(trainX,trainY)
y_pred_sklearn = KNN_Class.predict(testX)
```

```python
plt.plot(testY)
plt.show()
```

```python
plt.plot(y_pred_sklearn)
plt.show()
```

```python
plt.plot(testY)
plt.plot(y_pred_sklearn)
plt.show()
```

```python
accuracy_score(testY,y_pred_sklearn)*100
```
